```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
# use the 'seaborn-colorblind' style
plt.style.use('seaborn-colorblind')
```

```python
venues_df = pd.read_html("https://en.wikipedia.org/wiki/List_of_sports_venues_in_Australia")
venues_df = venues_df[0]
venues_df.rename(columns=venues_df.iloc[0], inplace=True)
venues_df.drop(venues_df.index[[0]], inplace=True)
venues_df.ix[2, 'Capacity'] = 75000
venues_df.ix[3, 'Capacity'] = 56347
venues_df.ix[9, 'Capacity'] = 43500
venues_df.ix[11, 'Capacity'] = 34074
venues_df.ix[14, 'Capacity'] = 27500
venues_df.ix[29, 'Capacity'] = 21000
venues_df.ix[33, 'Capacity'] = 19500
venues_df['Capacity'] = venues_df['Capacity'].astype(int)
```

```python
arenas_df = pd.read_html("https://en.wikipedia.org/wiki/List_of_indoor_arenas_in_Australia")
arenas_df = arenas_df[0]
arenas_df.rename(columns=arenas_df.iloc[0], inplace=True)
arenas_df.drop(arenas_df.index[[0]], inplace=True)
del arenas_df['Image']
arenas_df['State'] = arenas_df['Location'].map(lambda x: x.split(',')[1].strip())
arenas_df.ix[28, 'Capacity'] = 2000
arenas_df['Capacity'] = arenas_df['Capacity'].astype(int)
```

```python
arenas_capacity = arenas_df.groupby('State').sum()
arenas_capacity.ix['Northern Territory', 'Capacity'] = int(0)

venues_capacity = venues_df.groupby('State').sum()

arenas_capacity.sort_index(inplace=True)
venues_capacity.sort_index(inplace=True)

venues_capacity['State_Abbr'] = ['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA']
arenas_capacity['State_Abbr'] = ['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA']

venues_capacity.set_index('State_Abbr', inplace=True)
arenas_capacity.set_index('State_Abbr', inplace=True)
```

```python
sum_capacity = arenas_capacity['Capacity'] + venues_capacity['Capacity']
sum_capacity = sum_capacity.astype(int)
```

```python
def draw():
    plt.figure()
    bars = plt.bar(range(1, 9), venues_capacity['Capacity'], alpha=0.8)
    plt.bar(range(1, 9), arenas_capacity['Capacity'], bottom=venues_capacity['Capacity'], alpha=0.8)
    ax = plt.gca()

    # Set X tick labels
    xticklabels = [0] + list(arenas_capacity.index)
    ax.set_xticklabels(xticklabels);

    # Remove all the ticks
    ax.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off')

    # Remove all borders
    for (idx, spine) in enumerate(ax.spines.values()):
        spine.set_visible(False)

    # direct label each bar with Y axis values
    for idx, bar in enumerate(bars):
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, float(sum_capacity[idx] + 4000), str(sum_capacity[idx]), 
                     ha='center', fontsize=10, alpha=0.7)
    # Add a legend
    ax.legend(['Venues', 'Indoor Arenas'], frameon=False)

    # Add a title
    plt.title('Sports capacity in different states in Australia')

    # Save figure to file
    plt.savefig('Assignment4.png')
```
