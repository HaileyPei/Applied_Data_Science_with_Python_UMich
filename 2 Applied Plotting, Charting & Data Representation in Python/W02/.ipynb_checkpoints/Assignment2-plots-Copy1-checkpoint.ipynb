An NOAA dataset has been stored in the file `data/C2A2_data/BinnedCsvs_d100/2f0bb04162655f0cba429b865292f31482e817e0b3ee9da0f40185d7.csv`. The data for this assignment comes from a subset of The National Centers for Environmental Information (NCEI) [Daily Global Historical Climatology Network](https://www1.ncdc.noaa.gov/pub/data/ghcn/daily/readme.txt) (GHCN-Daily). The GHCN-Daily is comprised of daily climate records from thousands of land surface stations across the globe.

Each row in the assignment datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)
    * TMIN : Minimum temperature (tenths of degrees C)
* **value** : data value for element (tenths of degrees C)

For this assignment, you must:

Read the documentation and familiarize yourself with the dataset, then write some python code which returns a line graph of the record high and record low temperatures by day of the year over the period 2005-2014. The area between the record high and record low temperatures for each day should be shaded.
2. Overlay a scatter of the 2015 data for any points (highs and lows) for which the ten year record (2005-2014) record high or record low was broken in 2015.
3. Watch out for leap days (i.e. February 29th), it is reasonable to remove these points from the dataset for the purpose of this visualization.
4. Make the visual nice! Leverage principles from the first module in this course when developing your solution. Consider issues such as legends, labels, and chart junk.

The data you have been given is near **London, England, United Kingdom**, and the stations the data comes from are shown on the map below.

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(100,'2f0bb04162655f0cba429b865292f31482e817e0b3ee9da0f40185d7')
```

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.dates as dates
from matplotlib.ticker import NullFormatter
from matplotlib.dates import MonthLocator, DateFormatter
from pylab import figure, show, legend, ylabel


dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')
df = pd.read_csv("data/C2A2_data/BinnedCsvs_d100/2f0bb04162655f0cba429b865292f31482e817e0b3ee9da0f40185d7.csv", parse_dates=['Date'], date_parser=dateparse, index_col=False, usecols=['Date', 'Element', 'Data_Value'])
df.sort_values(['Date','Data_Value'], ascending=True, inplace=True)
df['Data_Value'] = df['Data_Value'].apply(lambda x: round(x*0.1)) #T in tenths of celsius to T in celsius 
df.set_index(['Date'], inplace=True)
df = df.groupby(pd.TimeGrouper('D')).Data_Value.agg({'TMIN': 'min', 'TMAX': 'max'}, inplace = True)
df = df.reset_index()
df['dayofyear'] = pd.DatetimeIndex(df['Date']).dayofyear.astype(np.int64)
df['month'] = pd.DatetimeIndex(df['Date']).dayofyear.astype(np.int64)
df['md'] = pd.DatetimeIndex(df['Date']).map(lambda x: x.strftime('%m-%d'))
df.set_index(['Date'], inplace=True)

def Feb_29(df):
    return (df.index.year % 4 == 0) & ((df.index.year % 100 != 0) | (df.index.year % 400 == 0)) & (df.index.month == 2) & (df.index.day == 29)
mask = Feb_29(df)
df[mask]
df = df[~mask] #Feb29 13 (2008) + 21 (2012) #length=4015

df1 = df['2015'] #length=365
df1.rename(columns={'TMAX': 'TMAX_2015', 'TMIN': 'TMIN_2015'}, inplace=True)
df1['ymd'] = np.arange('2015-01-01', '2016-01-01', dtype='datetime64[D]')

df2 = df['2005':'2014'] #length=3650
df2 = df2.reset_index(drop=True)
df2_min = df2.groupby(['md'], sort=True)['TMIN'].min()
df2_max = df2.groupby(['md'], sort=True)['TMAX'].max()
df1['TMIN_2005_2014'] = np.array(df2_min)
df1['TMAX_2005_2014'] = np.array(df2_max)

def max_hi_2015(df1):
    return (df1['TMAX_2015']>df1['TMAX_2005_2014'])
mask1 = max_hi_2015(df1)
df_hi_2015 = df1[mask1]

def min_lo_2015(df1):
    return (df1['TMIN_2015']<df1['TMIN_2005_2014'])
mask2 = min_lo_2015(df1)
df_lo_2015 = df1[mask2]

#print(df1.shape)
#print(df1.index)
#print(df1.columns)
#print(df1.info)
#print(df1.describe)
#print(df1.dtypes)
#df1.head(30)
# Create a figure size
fig= figure(figsize=(9, 7))
fig.subplots_adjust(left=0.12, right=0.8, bottom=0.2, top=0.8)

# Create line plots of the recorded high & low temperatures by day of the year over the period 2005-2014
ax = fig.add_subplot(111)

l1 = ax.plot(df1.index.values, df1['TMAX_2005_2014'], '-k', linewidth=1.1)
l2 = ax.plot(df1.index.values, df1['TMIN_2005_2014'], '-k', linewidth=0.4)

# Create Overlay scatterplots of the 2015 data for any points (highs & lows) for which the ten year record (2005-2014) record high or record low was broken in 2015
s1 = ax.scatter(df_hi_2015.index.values, df_hi_2015['TMAX_2015'], marker='o', color='r', s = 23)
s2 = ax.scatter(df_lo_2015.index.values, df_lo_2015['TMIN_2015'], marker='o', color='b', s = 23)

# Create legends
plt.legend([l1[0],l2[0],s1,s2], ['Max. T. (2005-2014)', 'Min. T. (2005-2014)', 'Higher T. (2015: n=29)', 'Lower T. (2015: n=8)'],loc='upper left', fontsize=9, frameon=False)

# Filling between line l1 and line l2
plt.fill_between(df1.index.values, df1.TMIN_2005_2014.values, df1.TMAX_2005_2014.values, facecolor='grey', alpha='0.15')

# Create dates range
date = np.arange('2015-01-01', '2016-01-01', dtype='datetime64[D]')
dates = list(map(pd.to_datetime, date))

ax = plt.gca()
# set dates limits
ax.set_xlim([dates[0], dates[-1]])
# formatters' options
ax.xaxis.set_major_locator(MonthLocator())
#ax.xaxis.set_minor_locator(MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(NullFormatter())
ax.xaxis.set_major_formatter(DateFormatter('%b'))
labels = ax.get_xticklabels()
plt.setp(labels, rotation=30, fontsize=11)
ax.set_xlabel('\n365 Days of Year (by Month)\n*Data excludes Feb-29 for 2008 & 2012', fontsize=11, fontweight='bold')
ax.set_ylabel('Temperature in Celsius ($^{\circ}$C)', fontsize=11, fontweight='bold')
ax.set_title('\nLine plot of Max. & Min. Temperature (2005-2014)\nvs.\nScatter plot of broken record of Max. & Min. Temperature (2015) in London (UK)\n', fontsize = 11, fontweight='bold')
#ax.set_axis_bgcolor('w')
# a degree Celsius to Fahrenheit - add second y-lables:
ymin, ymax = ax.get_ylim()
ymin_f = round(ymin*1.8 + 32).astype(np.int64) #0
ymax_f = round(ymax*1.8 + 32).astype(np.int64) #99
ax2= ax.twinx()
ax2.set_ylim(ymin_f, ymax_f)
ax2.set_ylabel('Temperature in Fahrenheit ($^{\circ}$F)', fontsize=11, fontweight='bold')

plt.autoscale(enable=True, axis='x', tight=True)
plt.show()
fig.savefig('Assignment2.png')
```
